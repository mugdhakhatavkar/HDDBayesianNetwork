# Convert bytes to TB

```
def bytes_to_TB(B):
   B = float(B)
   KB = float(1024)
   MB = float(KB ** 2) # 1,048,576
   GB = float(KB ** 3) # 1,073,741,824
   TB = float(KB ** 4) # 1,099,511,627,776
   return '{0:.2f}'.format(B/TB)
```

```
df["capacity_bytes"] = df["capacity_bytes"].apply(bytes_to_TB)
df
df["capacity_bytes"] = pd.to_numeric(df.capacity_bytes, errors='coerce',downcast='float')
df.drop(df[df['capacity_bytes'] > 8].index, inplace = True) 
```
